```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("shutterstock-2382173217.jpg")
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img,  cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title("BGR ")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(img2)
plt.title("RGB ")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(img3)
plt.title("Gray ")
plt.axis("off")

plt.show()
```

```python
import cv2

cap = cv2.VideoCapture("14503508_540_960_30fps.mp4")

while True:
    ret, frame = cap.read()

    if not ret:
        break

    cv2.imshow("Video", frame)

    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()

    if not ret:
        print("Failed to capture frame")
        break

    cv2.imshow("Webcam", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

img5 = cv2.imread("shutterstock-2382173217.jpg")

blur = cv2.blur(img5, (15, 15))

img5_rgb = cv2.cvtColor(img5, cv2.COLOR_BGR2RGB)
blur_rgb = cv2.cvtColor(blur, cv2.COLOR_BGR2RGB)

cv2.imshow("Original", img5)
cv2.imshow("Average Blur", blur)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(img5_rgb)
plt.title("Original (RGB)")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(blur_rgb)
plt.title("Average Blur")
plt.axis("off")

plt.tight_layout()
plt.show()

cv2.waitKey(0)
cv2.destroyAllWindows()
```
